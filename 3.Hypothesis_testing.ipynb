```python
import pandas as pd
from scipy import stats as stats
from scipy.stats import chi2_contingency
from scipy.stats import chi2
```

```python
from google.colab import files
uploaded=files.upload()
```

```python
data=pd.read_csv('/content/Costomer+OrderForm.csv')
data
```

```python
data.head()
```

```python
print(data['Phillippines'].value_counts(),data['Indonesia'].value_counts(),data['Malta'].value_counts(),data['India'].value_counts())
```

```python
observed=([[271,267,269,280]],[[29,33,31,20]])
```

```python
observed
```

```python
from scipy.stats.contingency import chi2_contingency
stat,p,dof,expected=chi2_contingency([[271,267,269,280],[29,33,31,20]])
```

```python
stat
```

```python
p
```

```python
print('dof=%d'%dof)
print(expected)
```

```python
alpha=0.05
prob=1-alpha
critical = chi2.ppf(prob,dof)
print('probability=%3f, critical=%3f, stats=%3f' % (prob,critical,stat))
if abs(stat) >= critical:
        print('Dependent (reject H0),variables are related')
else:
            print('Independent (fail to reject H0),variables are not related')
```

```python
print('significance=%.3f, p=%.3f' % (alpha,p))
if p <= alpha:
  print('Dependent (reject H0)')
else:
  print('Independent (fail to reject H0)')
```
