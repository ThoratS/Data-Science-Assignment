```python
rsq_rds = smf.ols('RDS~ADM+MKS',data=data1).fit().rsquared
vif_rds = 1/(1-rsq_rds)
vif_rds
```

```python
rsq_adm = smf.ols('ADM~RDS+MKS',data=data1).fit().rsquared
vif_adm = 1/(1-rsq_adm)
vif_adm
```

```python
rsq_mks = smf.ols('MKS~RDS+ADM',data=data1).fit().rsquared
vif_mks = 1/(1-rsq_mks)
vif_mks
```

```python
d1 = {'Variables':['RDS','ADM','MKS'],'VIF':[vif_rds,vif_adm,vif_mks]}
Vif_frame = pd.DataFrame(d1)
Vif_frame
```

```python
import statsmodels.api as sm
sns.set_style(style='darkgrid')
qqplot=sm.qqplot(model1.resid,line='q') # line = 45 to draw the diagnoal line
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model1.resid<-10000))
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(get_standardized_values(model1.fittedvalues),
            get_standardized_values(model1.resid))
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model1, "RDS", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model1, "ADM", fig=fig)
plt.show()
```

```python
from statsmodels.graphics.regressionplots import influence_plot

model1_influence = model1.get_influence()
(c, _) = model1_influence.cooks_distance
```

```python
sns.set_style(style='darkgrid')
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data1)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(15,10))
fig=influence_plot(model1,ax = ax)
```
